## Filtrando dados


```python
list(df_aluguel['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete', 
'Casa',
'Apartamento',
'Casa de Condomínio',
'Casa de Vila']
```

```python
df_aluguel['Tipo'].isin(residencial).head(10)
```

```python
selecao = df_aluguel['Tipo'].isin(residencial)
df_aluguel_residencial = df_aluguel[selecao]
df_aluguel_residencial.head(10)
```

```python
list(df_aluguel_residencial['Tipo'].drop_duplicates())
```

```python
df_aluguel_residencial.shape[0]
```

```python
df_aluguel_residencial.index = range(df_aluguel_residencial.shape[0])
df_aluguel_residencial
```
